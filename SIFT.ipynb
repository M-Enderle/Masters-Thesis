```python
from scipy.signal import stft
import matplotlib.pyplot as plt
import numpy as np
import scipy.io.wavfile as wav
import sys
import os
import json
```

```python
# read in all the training files
train_folder = "generated_training_data"

folders = [folder for folder in os.listdir(train_folder) if folder.startswith("soundscape")]
print(f"Found {len(folders)} soundscape folders")

train_dict = {}
for folder in folders:
    soundscape_path = os.path.join(train_folder, folder)
    train_dict[folder] = {}
    train_dict[folder]['wav'] = wav.read(os.path.join(soundscape_path, f"{folder}.wav"))
    train_dict[folder]['info'] = json.load(open(os.path.join(soundscape_path, f"{folder}_targets.json")))

train_dict
```
